```python
horses = pd.read_csv('/Users/maksim/Netology/material/практика_и_дз/horse_data.csv',na_values = '?')

#Ищем в файле horse_data.names описание столбцов и записываем названия в список
column_list =[]
with open("/Users/maksim/Netology/material/практика_и_дз/horse_data.names", "r") as f:
    for line in f:
        opisanie = re.findall(r'\d+:\s+([^\n]+)', line)
        
        if opisanie != []:
            opisanie = str(opisanie).replace("['","").replace("']","")
            column_list.append(opisanie)
        
        print(line)
        
f.close()

column_list.insert(25,'type of lesion')
column_list.insert(26,'type of lesion')
#column_list

```

```python
#Создание датафрейм new из двух столбцов
new = horses[['temperature of extremities', 'rectal temperature']]

#создание серии bool True для значений NaN
# фильтрация данных и отображение данных только с 'temperature of extremities' = NaN
bool_series = pd.isnull(new['temperature of extremities']) 
print(len(new[bool_series]))

bool_series1 = pd.isnull(new['rectal temperature']) 
print(len(new[bool_series1]))


#new[bool_series1]

```

```python
#напишем функцию которая будет заполнять пропуски в rectal temperature
#возьмем среднее значение для 
new.groupby('temperature of extremities').mean()

def search_rectal_temperature (par):
    if par == 1:
        return 38.126389
    elif par == 2:
        return 38.203846
    elif par == 3:
        return 38.183544
    elif par == 4:
        return 38.455000
    else:
        return mean_ #среднее для rectal_temperature    

#Можно написать похожую функцию для заполнения 'temperature of extremities'. Зная среднюю темпаретуру, указывать 1 из 4 значений. 
```

```python
#Добавляем столбец rectal temperature к основному датафрейм
df = new['rectal temperature'].fillna(new['temperature of extremities'].apply(search_rectal_temperature))
horses['rectal temperature'] = df
```

```python
print(horses['rectal temperature'].mean())
print(horses['rectal temperature'].median())
print(horses['rectal temperature'].mode())
```
